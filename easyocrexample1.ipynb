```
# importing & reading easyocr  
import easyocr
reader = easyocr.Reader(['en'])
```

```
# showing an image
import PIL
from PIL import ImageDraw
img = PIL.Image.open("Image.jpg")
img
```

```
# Getting bounding boxes
bounds = reader.readtext('Image.jpg')
bounds
```

```
# Drawing bounding boxes
def draw_boxes(image, bounds, color='red', width=1):
    draw = ImageDraw.Draw(image)
    for bound in bounds:
        pix0, pix1, pix2, pix3 = bound[0]
        draw.line([*pix0, *pix1, *pix2, *pix3, *pix0], fill=color, width=width)
    return image

draw_boxes(img, bounds)

#len(bounds)
```
