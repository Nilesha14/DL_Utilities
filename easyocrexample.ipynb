```
# importing easyocr
import easyocr
```

```
#en - English, hi - Hindi, mr - Marathi
reader = easyocr.Reader(['en', 'hi', 'mr'])
```

```
# importing urllib
import urllib.request
```

```
# specify the image file path of image and retriving image

#English 
strImg='https://i1.wp.com/9to5google.com/wp-content/uploads/sites/4/2020/03/pixel_4a_blass_billboard_2.jpg'

#Hindi
#strImg='https://www.achisoch.com/wp-content/uploads/2016/05/Achi-Soch-Quotes-on-Life-in-Hindi.jpg'

#Marathi
#strImg='https://i.ytimg.com/vi/3OsD2Ugj6gU/maxresdefault.jpg'

#Eng,Hindi and Marathi
#strImg='https://images-eu.ssl-images-amazon.com/images/I/51jLLtU4y-L._SX300_QL70_ML2_.jpg'

urllib.request.urlretrieve(strImg, filename = 'inputfile.jpg')
```
